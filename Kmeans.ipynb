```python
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
```

```python
import pandas as pd
iris = load_iris()
df = pd.DataFrame(iris["data"], columns=iris['feature_names'])
df['target'] = iris['target']
df
```

```python
# 切開來 -> 為之後步驟做準備
from sklearn.model_selection import train_test_split
 #[90%Question, 10%Question, 90%Question, 10%Answer]
x_train, x_test, y_train, y_test = train_test_split(iris['data'], iris['target'], test_size=0.1)
```

```python
clu = KMeans(n_clusters=2)
clu.fit(x_train)
```

```python
print('預測', list(clu.predict(x_test)))
```

```python
print('本來的標籤', list(y_test))
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(clu.predict(x_test), y_test)
```

```python
# 將所有標籤取出
clu.labels_
```

```python
from sklearn.metrics import silhouette_score
xs = []
ys = []
for k in range(2, 6):
    clu = KMeans(n_clusters=k)
    clu.fit(iris['data'])
    xs.append(k)
    score = silhouette_score(iris['data'], clu.labels_)
    ys.append(score)
print(xs)
print(ys)
```

```python
import matplotlib.pyplot as plt
#https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html
%matplotlib inline
plt.plot(xs, ys, 'm^--')
plt.title('Sihouette Score')
plt.xlabel('K')
plt.ylabel('Score')
#plt.show()
```
