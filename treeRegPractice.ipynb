```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.datasets import load_boston
```

```python
import pandas as pd
boston = load_boston()
df = pd.DataFrame(boston["data"], columns=boston['feature_names'])
df['target'] = boston['target']
df
```

```python
# 切開來 -> 為之後步驟做準備
from sklearn.model_selection import train_test_split
 #[90%Question, 10%Question, 90%Question, 10%Answer]
x_train, x_test, y_train, y_test = train_test_split(boston['data'], boston['target'], test_size=0.1)
train_test_split(boston['data'], boston['target'], test_size=0.1)
```

```python
# (90%) (訓練)算出方程式
reg = DecisionTreeRegressor(max_depth=5)
reg.fit(x_train, y_train)
```

```python
from sklearn.tree import export_graphviz
import graphviz
g = export_graphviz(reg, feature_names=boston["feature_names"], filled=True)
graphviz.Source(g)
```

```python
from sklearn.metrics import accuracy_score
# (10%) (測試) 驗證方程式好不好
print('預測' , list(reg.predict(x_test)))
print('正確' , list(y_test))
```
