```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.datasets import load_boston
```

```python
import pandas as pd
boston = load_boston()
df = pd.DataFrame(boston["data"], columns=boston['feature_names'])
df['target'] = boston['target']
df
```

```python
# åˆ‡é–‹ä¾† -> ç‚ºä¹‹å¾Œæ­¥é©Ÿåšæº–å‚™
from sklearn.model_selection import train_test_split
 #[90%Question, 10%Question, 90%Question, 10%Answer]
x_train, x_test, y_train, y_test = train_test_split(boston['data'], boston['target'], test_size=0.1)
train_test_split(boston['data'], boston['target'], test_size=0.1)
```

```python
# (90%) (è¨“ç·´)ç®—å‡ºæ–¹ç¨‹å¼
reg = DecisionTreeRegressor(max_depth=6)
reg.fit(x_train, y_train)
```

```python
from sklearn.tree import export_graphviz
import graphviz
g = export_graphviz(reg, feature_names=boston["feature_names"], filled=True)
graphviz.Source(g)
```
